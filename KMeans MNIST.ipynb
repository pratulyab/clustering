# KMeans on MNIST

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('/Users/Pratulya/Desktop/machine-learning/bootcamp/mnist_data/train.csv')
data = dataset.values
print(data.shape)
```

```python
# data_till_5 = data[data[:,0] <= 5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[:, 1:], data[:, 0], shuffle=True)
# X_train, X_test, y_train, y_test = train_test_split(data_till_5[:, 1:], data_till_5[:, 0], test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
k = 100
kmeans = KMeans(n_clusters=k)
kmeans
```

```python
# kmeans.fit(X_train)
```

```python
# kmeans.labels_[:100]
```

```python
cluster_labels = kmeans.fit_predict(X_train)
cluster_labels.shape
```

```python
X_train.shape
```

```python
def analyse_labels(cid):
    plt.figure()
#     mean_digit = ((X_train[cluster_labels == cid]).reshape((-1, 28, 28))).mean(axis=0)
    mean_digit = kmeans.cluster_centers_[cid].reshape((28,28))
    plt.imshow(mean_digit)
    plt.show()
```

```python
cid_to_digit = {}
for i in range(k):
    analyse_labels(i)
    digit = int(input("Enter Digit: "))
    cid_to_digit[i] = digit
cid_to_digit
```

```python
# plt.figure()
# plt.imshow(X_train[cluster_labels == 6][434].reshape((28,28)))
```

```python
print(cluster_labels[:10])
```

```python
cluster_labels = kmeans.labels_.copy()
for i in range(len(cluster_labels)):
    cluster_labels[i] = cid_to_digit[cluster_labels[i]]
print(cluster_labels[:10])
```

```python
score = (cluster_labels == y_train).sum() / y_train.shape[0]
score
```

### Testing on Test Set

```python
pred = kmeans.predict(X_test)
for i in range(len(pred)):
    pred[i] = cid_to_digit[pred[i]]
print("Score: ", (pred == y_test).sum()/pred.shape[0])
```
